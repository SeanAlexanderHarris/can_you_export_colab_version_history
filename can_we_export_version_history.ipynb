### Random German Name Generator

This notebook will select and display a random German name from a list of 20 names.

```python
import random

# List of 20 German names
german_names = [
    "Anna", "Max", "Sophie", "Felix", "Mia", "Luca", "Hannah", "Paul", "Lena", "Jonas",
    "Emilia", "Leon", "Marie", "David", "Laura", "Noah", "Clara", "Ben", "Julia", "Tim"
]

# Select a random name from the list
random_name = random.choice(german_names)

# Display the randomly selected name
print(f"Randomly generated German name: {random_name}")
```
